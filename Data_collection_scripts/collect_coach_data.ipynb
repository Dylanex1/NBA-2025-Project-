```python
import pandas as pd

url = "https://www.basketball-reference.com/leagues/NBA_2025_coaches.html"

tables = pd.read_html(url)  

df = tables[0]   # Extracting only table       

df = df.drop(df.columns[[2, 5, 16]], axis=1)

df.columns = [
    "Coach", "Tm", 
    "Seasons_Franchise", "Seasons_Overall",
    "Reg_Current_G", "Reg_Current_W", "Reg_Current_L",
    "Reg_Franchise_G", "Reg_Franchise_W", "Reg_Franchise_L",
    "Reg_Overall_G", "Reg_Overall_W", "Reg_Overall_L", "Reg_Overall_W%",
    "Playoffs_Current_G", "Playoffs_Current_W", "Playoffs_Current_L",
    "Playoffs_Franchise_G", "Playoffs_Franchise_W", "Playoffs_Franchise_L",
    "Playoffs_Overall_G", "Playoffs_Overall_W", "Playoffs_Overall_L"
]

df.to_csv("coaches_24_25.csv", index=False)
print(df.head())
```
