```python
import pandas as pd

game = pd.read_csv("../Stage_3/Datasets/Game.csv")
team = pd.read_csv("../Stage_3/Datasets/team.csv")
```

```python
# Remove result
game = game.drop(columns = ["Result"])
```

```python
# Replace team names with ids
team_to_id = {
    team : id for team, id in zip(team["Team name"], team["Team_id"])
}

home_ids = []
visitor_ids = []

for home_team, visitor_team in zip(game["home team name"], game["Visitor team name"]):
    home_ids.append(team_to_id[home_team])
    visitor_ids.append(team_to_id[visitor_team])

game["HomeTeamID"] = home_ids
game["VisitorTeamID"] = visitor_ids
```

```python
# Sanity Checks
print("Unmapped home teams:", game["HomeTeamID"].isna().sum())
print("Unmapped visitor teams:", game["VisitorTeamID"].isna().sum())
```

```python
game = game.drop(columns = ["home team name", "Visitor team name"])
game.to_csv("../Stage_3/Datasets/bcnf_game.csv", index = False)
```
