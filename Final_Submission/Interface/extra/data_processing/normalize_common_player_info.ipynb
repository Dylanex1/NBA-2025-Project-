```python
import pandas as pd

player_info = pd.read_csv("../Stage_3/Datasets/common_player_info.csv")
player = pd.read_csv("../Stage_3/Datasets/player.csv")
```

```python
# Month to inches map
mon_map = {"Jan" : 1, "Feb" : 2, "Mar" : 3, "Apr" : 4, "May" : 5, "Jun" : 6, "Jul" : 7, "Aug" : 8}

# convert date to height
heights = []
for date in player_info["height"]:
    height = pd.NA
    if not pd.isna(date):
        parts = date.split("-")
        feet = mon_map.get(parts[1])
        if feet is None:
            feet = int(parts[0])
            inches = int(parts[1])
        else:
            inches = int(parts[0])

        height = str(feet) + "\"" + str(inches)

    heights.append(height)
```

```python
# Sanity checks
print("Length of new heights:", len(heights))
print("Length of old heights:", len(player_info["height"]))
print("Number of NAs in old heights:", pd.isna(player_info["height"]).sum())
print("Number of NAs in new heights:", pd.isna(heights).sum())
```

```python
# Replace old heights with new heights
player_info["height (ft\"in)"] = heights
player_info = player_info.drop(columns = ["height"])

player_info['birthdate'] = pd.to_datetime(player_info["birthdate"]).dt.date
```

```python
# Drop unnecessary columns
# Roster status is derived from the player table
player_info = player_info.drop(columns = ["roster status"])
```

```python
id_to_active = {
    id : active for id, active in zip(player["PlayerID"], player["is_active"])
}

is_active = []
for id in player_info["person_id"]:
    is_active.append(id_to_active[id])

player_info["is_active"] = is_active
player = player.drop(columns = ["is_active"])
```

```python
player_info = player_info.rename(columns = {'person_id' : 'PlayerID', 'country' : 'BirthCountry'})
player_info.to_csv("../Stage_3/Datasets/bcnf_player_info.csv", index = False)
player.to_csv("../Stage_3/Datasets/bcnf_player.csv", index = False)
```
